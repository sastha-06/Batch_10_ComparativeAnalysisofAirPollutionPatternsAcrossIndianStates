```python
import pandas as pd
import os
# Make sure the reports directory exists
os.makedirs("../reports", exist_ok=True)

df = pd.read_csv("city_day.csv")

# Missing values
missing_values = df.isnull().sum()

# Duplicate rows
duplicate_rows = df.duplicated().sum()

print("Missing Values:\n", missing_values)
print("\nDuplicate Rows:", duplicate_rows)

# Save Data Quality Report (RULE BOOK MANDATORY)
with open("../reports/data_quality_report.txt", "w") as f:
    f.write("DATA QUALITY REPORT - CITY DATASET\n\n")
    f.write("Missing Values per Column:\n")
    f.write(missing_values.to_string())
    f.write("\n\nDuplicate Rows: " + str(duplicate_rows))

```
